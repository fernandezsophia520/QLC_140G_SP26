```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# creating figure 10x5 inch
fig= plt.figure (figsize=(10,5))
ax= fig.add_subplot(1,1,1, projection=ccrs.PlateCarree())
ax.coastlines()

# Coordinates for Charlotte, North Carolina (from kernel state)
charlotte_lon = -80.8431
charlotte_lat = 35.2271

# Add text label 'we are here'
ax.text(charlotte_lon + 3, charlotte_lat + 3, 'we are here', transform=ccrs.PlateCarree(),
        horizontalalignment='left', color='red', fontsize=12, weight='bold')

# Add major ocean labels
# Atlantic Ocean
ax.text(-30, 0, 'Atlantic Ocean', transform=ccrs.PlateCarree(),
        horizontalalignment='center', color='blue', fontsize=12, weight='bold')
# Pacific Ocean (left side)
ax.text(-160, 0, 'Pacific Ocean', transform=ccrs.PlateCarree(),
        horizontalalignment='center', color='blue', fontsize=12, weight='bold')
# Pacific Ocean (right side, for wraparound)
ax.text(160, 0, 'Pacific Ocean', transform=ccrs.PlateCarree(),
        horizontalalignment='center', color='blue', fontsize=12, weight='bold')
# Indian Ocean
ax.text(90, -20, 'Indian Ocean', transform=ccrs.PlateCarree(),
        horizontalalignment='center', color='blue', fontsize=12, weight='bold')
# Southern Ocean
ax.text(0, -70, 'Southern Ocean', transform=ccrs.PlateCarree(),
        horizontalalignment='center', color='blue', fontsize=12, weight='bold')

# Add latitude lines with labels and grid
grid_lines = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=True,
                           linewidth=1, color='gray', alpha=0.5, linestyle='--')

# Set major latitude ticks for the gridlines and labels
grid_lines.ylocator = mticker.FixedLocator(np.arange(-90, 91, 30))

# Remove longitude labels from the top and bottom sides, and latitude labels from the right side
grid_lines.xlabels_top = False
grid_lines.xlabels_bottom = False
grid_lines.ylabels_right = False

plt.title ('Map of Earth with Charlotte, NC, Ocean Labels, and Latitude Grid')
plt.show()
```
