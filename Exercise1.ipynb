This exercise will introduce plython programming and plotting nutrient data usng phython package called matplotlib

```python
#First we create a cell that installs packages we'll be uisng
import pandas as pd
```

```python
#start by importing the nutrient data by hand:
data= {
  "Depth_m": [0,50,100,200,500,1000,2000,4000],
  "Nitrate_Atlantic_uM":[0.5,1.2,3.0,8.5,18.0,24.0,30.0,32.0],
  "Nitrate_Pacific_uM":[0.6,2.0,6.0,15.0,28.0,36.0,42.0,45.0]
}
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['Nitrate_Atlantic_uM'], df['Depth_m'])
plt.xlabel('Nitrate_Atlantic_uM')
plt.ylabel('Depth_m')
plt.title('Atlantic Nitrate vs. Depth')
plt.gca().invert_yaxis() # Invert y-axis to show depth increasing downwards
plt.grid(True)
plt.show()
```

```python
#do the same with the pacific
#Creating a figure using matplotlib
plt.figure(figsize=(8,6))
#Now plot a scatter plot in this empty figure
plt.scatter(df['Nitrate_Pacific_uM'], df['Depth_m'])
#Create a X-axis label describing what is on the axis
plt.xlabel('Nitrate_Pacific_uM')
#Create a x-axis label decribing what is on the axis
plt.ylabel('Depth_m')
#Create a figure title
plt.title('Nitrate profile in the pacific ocean')
plt.gca().invert_yaxis() # Invert y-axis so 0 depth is at the top
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))

# Plot Atlantic Nitrate
plt.plot(df['Nitrate_Atlantic_uM'], df['Depth_m'], label='Nitrate_Atlantic_uM', marker='o')

# Plot Pacific Nitrate
plt.plot(df['Nitrate_Pacific_uM'], df['Depth_m'], label='Nitrate_Pacific_uM', marker='x')


plt.xlabel('Nitrate (uM)')
plt.ylabel('Depth (m)')
plt.title('Nitrate Profiles: Atlantic vs. Pacific')
plt.gca().invert_yaxis() # Invert y-axis to show depth increasing downwards
plt.legend()
plt.grid(True)
plt.show()
```
